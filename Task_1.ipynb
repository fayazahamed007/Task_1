```python
percent = (df.isnull().sum()/df.shape[0] * 100).round(2)
percent
```

```python
missing_percent  = 80
index_clean = percent[percent < missing_percent].index
df_updated_data = df[index_clean].copy()
```

```python
df_updated_data.head()
```

```python
df_updated_data.loc[:,'director'] = df_updated_data['director'].fillna('Not available')
df_updated_data.loc[:,'cast'] = df_updated_data['cast'].fillna('Not available')
df_updated_data.loc[:,'rating'] = df_updated_data['rating'].fillna('Not available')
df_updated_data.loc[:,'country'] = df_updated_data['country'].fillna('Not available')
df_updated_data.loc[:,'date_added'] = df_updated_data['date_added'].fillna('Not available')
df_updated_data.loc[:,'duration'] = df_updated_data['duration'].fillna('Not available')
```

```python
df_updated_data
```

```python
df_updated_data.isnull().sum()
```

```python
df = df_updated_data
```

```python
cat_df = df.select_dtypes(include=['object']).columns.to_list()
num_df = df.select_dtypes(include=['int']).columns.to_list()
print("Categorical Variables:",format(cat_df))
print("Numerical Variables:",format(num_df))
```

```python
# Convert 'date_column' to datetime
df['date_added'] = pd.to_datetime(df['date_added'],errors='coerce')
df['year'] = df['date_added'].dt.year
print(df)
```

```python
#Seaborn
import seaborn as sns
plt.figure(figsize=(12,6))
sns.countplot(data=df,x='type')
plt.xticks(rotation=45);
```

```python
import matplotlib.pyplot as plt
pie_chart = df['type'].copy().value_counts()
plt.figure(figsize=(6,6))
plt.pie(x=pie_chart.values,labels=pie_chart.index,wedgeprops={'linewidth':2.0,'edgecolor':'white'})
plt.title('Movie and TV Shows Ratio')
plt.legend(labels=[f"{label} ({count})" for label, count in zip(pie_chart.index, pie_chart.values)],loc='best')
plt.show()
```

```python
#We group the top_4 country
country_top4 = df[(df["country"]=="United States") | (df["country"]=="India") | (df["country"]=="United Kingdom") | (df["country"]=="Canada")]
country_top4.head(10)
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="country" , hue="type" , data=country_top4, palette="Set2")
plt.title("Comparison of content type available in top 4 countries", fontsize=12)
```

```python
plt.figure(figsize=(20,8))
sns.set_theme(style = "whitegrid")
sns.stripplot(x="country" , y="rating" , hue="type", data=country_top4 , order=["United States" , "India" , "United Kingdom" , "Canada"])
plt.legend(loc="upper left" , title="Type" , fontsize=12)
plt.title("Distribution of rating-based movies and shows in top 4 countries", fontsize=15)
```
